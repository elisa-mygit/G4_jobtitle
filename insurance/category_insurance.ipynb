```python
import pickle
import numpy as np
import os

def dummy_f(doc):
    return doc

class Category():

    def __init__(self, path=os.getcwd() + r"\insurance_pred.pkl"):
        with open(path, 'rb') as f:
            self.model = pickle.load(f)

    def Get_Category(self, x):
        model = self.model
        if type(x) is list:
            pass
        else:
            x=[x]
        cat = model.predict(x)
        cat = np.array2string(cat)
        cat = cat[2:-2]
        return cat
```

```python
classifier = Category()
ex1 = "核廢料處理人員"
cat = classifier.Get_Category(ex1)
print("職稱: {} \n分類: {}".format(ex1, cat))
```
